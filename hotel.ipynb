```python
import random

class People:
    def __init__(self):
        self.human = random.choice(["Бізнесмен", "Спортсмен", "Людина"])
        self.age = random.randint(18, 90)
        self.comfort = 0
        self.stress = 0
        self.kids = 0
        self.business = 0
        self.hunger = 0
        self.money = 0
        self.room = random.choice(["Лайт", "Важкий люкс"])

class HumanMoney:
    def bank(self, person):
        if person.human == "Бізнесмен":
            person.money += 10000
        elif person.human == "Спортсмен":
            person.money += 5000
        elif person.human == "Людина":
            person.money += 2500


class HotelClassic:
    def classic_service(self, person):
        self.pool = random.randint(1, 6)  
        self.eat = random.randint(1, 6)
        self.sing = random.randint(1, 6)
        if person.human == "Людина":
            if self.pool >= 2:
                person.money -= 150
                person.comfort += 1
                person.stress = 0
            if self.eat >= 2:
                person.money -= 350
                person.comfort += 1
                person.hunger += 1
                person.stress = 0
            if self.sing >= 2:
                person.comfort += 1
                person.money -= 100
                person.stress = 0
            if person.kids >= 2:
                person.money -= 100
            if person.room == "Лайт":
                person.money -= 300
                person.comfort += 1
            if person.room == "Важкий люкс":
                person.money -= 400
                person.comfort += 1    

class HotelSport:
    def sport_service(self, person):
        self.pool = random.randint(1, 6)  
        self.eat = random.randint(1, 6)
        self.sing = random.randint(1, 6)
        self.boxing = random.randint(1, 6)
        self.volleyball = random.randint(1, 6)
        self.football = random.randint(1, 6)
        self.basketball = random.randint(1, 6)
        if person.human == "Спортсмен":
            if self.pool >= 2:
                person.stress = 0
                person.money -= 200
                person.comfort += 1
            if person.room == "Лайт":
                person.money -= 1200
                person.comfort += 1
            if person.room == "Важкий люкс":
                person.money -= 2000
                person.comfort += 1
            if self.eat >= 2:
                person.money -= 450
                person.comfort += 1
                person.hunger += 1
            if self.sing >= 2:
                person.comfort += 1
                person.money -= 100
            if person.kids >= 2:
                person.money -= 100 
            if self.volleyball >= 2:
                person.stress = 0
                person.money -= 190
            if self.football >= 2:
                person.stress = 0
                person.money -= 170
            if self.basketball >= 2:
                person.stress = 0
                person.money -= 160

class HotelBusiness:
    def business_service(self, person):
        self.pool = random.randint(1, 6)
        self.eat = random.randint(1, 6)
        self.drink = random.randint(1, 6)
        self.sing = random.randint(1, 6)
        self.golf = random.randint(1, 6)
        if person.human == "Бізнесмен":
            if self.pool >= 2:
                person.money -= 400
                person.comfort += 1
                person.stress = 0
            if self.eat >= 2:
                person.money -= 350
                person.comfort += 1
                person.hunger += 1
                person.stress = 0
            if self.sing >= 2:
                person.comfort += 1
                person.money -= 100
                person.stress = 0
            if person.kids >= 2:
                person.money -= 100
            if person.room == "Лайт":
                person.money -= 2500
                person.comfort += 1
            if person.room == "Важкий люкс":
                person.money -= 3500
                person.comfort += 1

# Решта коду залишається незмінною

person = People()
money = HumanMoney()
money.bank(person)
classic = HotelClassic()
classic.classic_service(person)
sport = HotelSport()
sport.sport_service(person)
busines = HotelBusiness()
busines.business_service(person)

print(f"ОСОБА: {person.human}")
print(f"Вік: {person.age}")
print(f"Стрес після отримання послуг: {person.stress}")
print(f"Баланс після отримання послуг: {person.money}")
print(f"Комфорт людини після отримання послуг: {person.comfort}")
if person.human == "Людина":
    print(f"Обраний номер для людини: {person.room}")
elif person.human == "Спортсмен":
    print(f"Обраний номер для спортсмена: {person.room}")
elif person.human == "Бізнесмен":
    print(f"Обраний номер для бізнесмена: {person.room}")

```
